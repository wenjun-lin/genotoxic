```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
matplotlib.style.use('ggplot')
```

```python
data = pd.read_csv('toxic_data.csv')
plate_list = set(data['PlateName'])
```

```python
for name in plate_list:
    data_plate = data[data['PlateName']==name]
    feature_choose1 = 'Micronuclei_-_Number_of_Objects'
    negative = data_plate[data_plate['Compound_Name']=='DMSO'][feature_choose1].mean()
    sns.stripplot(data_plate['Class'],np.log2(data_plate[feature_choose1]/negative),size=1,jitter=True)
```
